```python
def process_query(query, index):
    """
    Process a query string and return the documents that match the query.
    """
    query_parts = query.lower().split()
    result = set(index.get(query_parts[0], []))
    for i in range(1, len(query_parts), 2):
        if query_parts[i] == 'ู':
            result &= set(index.get(query_parts[i+1], []))
        elif query_parts[i] == 'ุฃู ':
            result |= set(index.get(query_parts[i+1], []))
        elif query_parts[i] == 'ูููุณ':
            result -= set(index.get(query_parts[i+1], []))
    return result
```

```python
def evaluate(queries, index, preprocessed_documents):
    """
    Evaluate the performance of the index by computing precision and recall for each query.
    """
    evaluation = []
    for query in queries:
        # Get the documents that match the query
        retrieved = process_query(query, index)
        relevant = set()
        # Compute the number of relevant documents
        for i, doc in enumerate(preprocessed_documents):
            if any(word in doc for word in query.split()):
                relevant.add(i)
        # Compute precision and recall
        precision = len(retrieved.intersection(relevant)) / len(retrieved) if retrieved else 0
        recall = len(retrieved.intersection(relevant)) / len(relevant) if relevant else 0
        evaluation.append((query, precision, recall))
    return evaluation
```

**The following implementation works for both phrase queries and wildcard queries**


```python
import re

# Define a function to preprocess text
def preprocess(text):
    # Convert to lowercase
    text = text.lower()
    # Remove non-alphanumeric characters
    text = re.sub(r'\W+', ' ', text)
    # Split into words
    words = text.split()
    return words

def build_index(docs):
    index = {}
    for i, doc in enumerate(docs):
        # Split the document into words
        words = doc.split()
        # Iterate over the words and update the index
        for j, word in enumerate(words):
            # Add the document to the index for the current word
            if word not in index:
                index[word] = {}
            if i not in index[word]:
                index[word][i] = []
            index[word][i].append(j)
    return index


def search(query, index, docs):
    # Check if the query is a wildcard query
    if "*" in query:
        # Split the query into prefix and suffix
        prefix, suffix = query.split("*")
        # Initialize the result set
        result = set()
        # Iterate over the words in the index
        for word in index:
            # Check if the word has the prefix and suffix
            if word.startswith(prefix) and word.endswith(suffix):
                # Get the postings for the word and add them to the result set
                result.update(index[word].keys())
    else:
        # Split the query into words
        query_words = query.lower().split()
        # Initialize the result set
        result = set()
        # Get the postings for the first word in the query
        word = query_words[0]
        postings = index.get(word, {})
        # Iterate over the documents in the postings for the first word
        for doc_id, positions in postings.items():
            # Check if the document contains the query
            found = True
            # Iterate over the positions of the first word in the document
            for pos in positions:
                # Check if the remaining words in the query are in the same document and in the correct order
                for i in range(1, len(query_words)):
                    next_word = query_words[i]
                    next_positions = index.get(next_word, {}).get(doc_id, [])
                    if not next_positions or not any(next_pos == pos+i for next_pos in next_positions):
                        found = False
                        break
                if found:
                    result.add(doc_id)
    # Return the documents that match the query
    return [docs[i] for i in result]
```

```python
docs = [
    "the quick brown fox jumps over the lazy dog",
    "the lazy dog jumps over the quick brown fox",
    "the quick brown fox is quick",
    "the lazy dog is lazy",
    "the quick fox is brown"
]

index = build_index(docs)

# Test phrase query
query = "quick brown fox"
results = search(query, index, docs)
print("Results for phrase query '{}':".format(query))
print(results)
print('*' * 100)
print('*' * 100)
# Test wildcard query
query = "laz*"
results = search(query, index, docs)
print("Results for wildcard query '{}':".format(query))
print(results)
```
