```python
import os
import re
import string
import itertools
import pandas as pd
from string import digits
import pyarabic.araby as araby
from snowballstemmer import stemmer
from sklearn.feature_extraction.text import CountVectorizer
```

```python
documents = []

for file in os.listdir("/content/sample_data/data"):
    if file.endswith(".txt"):
        doc = os.path.join("/content/sample_data/data", file)
        documents.append(doc)
        print(doc)
        
```

---------------------------

<br>
<div class="alert alert-warning">
<b>Stemming</b>     
</div>


```python
ar_stemmer = stemmer("arabic")
stem = ar_stemmer.stemWord(u"/content/sample_data/data")
stem
```

---------------------------

<br>
<div class="alert alert-warning">
<b>Data before cleaning</b>     
</div>

```python
data_before_cleaning = [open(doc, encoding="utf8").read() for doc in documents]

data_before_cleaning
```

---------------------------

<br>
<div class="alert alert-warning">
<b>Remove new line</b>     
</div>

```python
documents_data = [re.sub('\n' , ' ' , doc) for doc in data_before_cleaning]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>remove english characters</b>     
</div>

```python
documents_data = ["".join([char for char in doc if char not in string.ascii_letters]).strip() for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>remove emojis</b>     
</div>

```python
emoji_pattern = re.compile(pattern = "["    u"\U0001F600-\U0001F64F"  
    u"\U0001F300-\U0001F5FF"  
    u"\U0001F680-\U0001F6FF"  
    u"\U0001F1E0-\U0001F1FF"  
    u"\U00002500-\U00002BEF"  
    u"\U00002702-\U000027B0"
    u"\U00002702-\U000027B0"
    u"\U000024C2-\U0001F251"
    u"\U0001f926-\U0001f937"
    u"\U00010000-\U0010ffff"
    u"\u2640-\u2642" 
    u"\u2600-\u2B55"
    u"\u200d"
    u"\u23cf"
    u"\u23e9"
    u"\u231a"
    u"\ufe0f"
    u"\u3030""]+", flags = re.UNICODE)

documents_data = [emoji_pattern.sub(r'',doc) for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>replace consuctive spaces with onle one</b>     
</div>

```python
documents_data = [' '.join(doc.split()) for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>remove repeated characters</b>     
</div>

```python
documents_data = [''.join(i for i, _ in itertools.groupby(doc)) for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>remove diacritics</b>     
</div>

```python
documents_data = [araby.strip_diacritics(doc) for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>remove punctuations</b>     
</div>

```python
documents_data = [doc.translate(str.maketrans('', '', string.punctuation)) for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>normalization (remove ุงู)</b>     
</div>

```python
documents_data = [re.sub('ุงู' , '' , doc) for doc in documents_data]
documents_data
```

---------------------------

<br>
<div class="alert alert-warning">
<b>compare between sample document before and after cleaning</b>     
</div>

```python
print('Before Cleaning: ',data_before_cleaning[0])
print('After Cleaning: ',documents_data[0])
```

```python
countvectorizer = CountVectorizer()
BOW_representation = countvectorizer.fit_transform(documents_data).toarray()
BOW_representation
```
